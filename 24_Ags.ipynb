```python
# Exercise 1

# Menyimpan dan Menampilkan Data (ga disuruh input)

nama1 = "Kopi Pagi"
nama2 = "Roti Cokelat"
harga_kopi = 18000.5
harga_roti = 10000
status_kopi = True

print("Nama produk 1:", nama1)
print("Harga produk 1:", harga_kopi)
print("Nama produk 2:", nama2)
print("Harga produk 2:", harga_roti)
print("Status ketersediaan roti:", status_kopi)
```

```python
# Exercise 2

# Konversi Tipe Data dan Input Pengguna

_kopi_str = input("Masukkan jumlah pesanan kopi: ")
_roti_str = input("Masukkan jumlah pesanan roti: ")

print("Masukkan jumlah pesanan kopi:", _kopi_str)
print("Masukkan jumlah pesanan roti:", _roti_str)
print("Tipe data awal jumlah kopi:", type(_kopi_str))
print("Tipe data awal jumlah roti:", type(_roti_str))

_kopi_str = int(_kopi_str)
_roti_str = int(_roti_str)

# boleh salah satu

print("Tipe data setelah konversi:", type(_kopi_str))
```

```python
# Exercise 3

# Operasi Pada Angka

total_kopi = harga_kopi * _kopi_str
totalRoti = harga_roti * _roti_str

print("Total harga kopi:", total_kopi)
print("Total harga roti:", totalRoti)

total_belanja = total_kopi + totalRoti
print("Total belanja keseluruhan:", total_belanja)

uang_bayar = 50000
print("Uang yang dibayarkan:", uang_bayar)

kembalian = uang_bayar - total_belanja
print("Kembalian:", kembalian)
```
